```python
import numpy as np
import pandas as pd
from datetime import timedelta
```

```python
df = pd.read_csv('Cleaned_NFL_Data.csv')
df.head()
```

```python
#convert Possession into Seconds
def timefinder(x):
    mins,secs = map(float, x.split(':'))
    td = timedelta(minutes=mins, seconds=secs)
    return td.total_seconds()
df['possessioninseconds_away'] = df['possession_away'].apply(timefinder)
df['possessioninseconds_home'] = df['possession_home'].apply(timefinder)

df.head()
```

```python
df = df.drop(columns=['possession_away', 'possession_home'])
df.head()
```

```python
df.to_csv("NFL_VIZ_Data.csv",index=0)
```
